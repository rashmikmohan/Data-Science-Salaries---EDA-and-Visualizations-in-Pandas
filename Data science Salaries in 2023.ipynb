```python
df['experience_level']=df['experience_level'].replace(['SE','EX','MI','EN']
                                                      ,['Senior Level','Executive Level','Medium Level','Entry Level'])
```

```python
df['employment_type']=df['employment_type'].replace(['FT','CT','PT','FL'],['Full Time','Contract','Part Time','Freelance'])
```

```python
df['remote_ratio'] = df['remote_ratio'].replace([0, 50, 100], 
                                                ['On-Site', 'Hybrid', 'Remote'])
```

```python
df['company_size'] = df['company_size'].replace(['S', 'M', 'L'], 
                                                ['Small', 'Medium', 'Large'])
```

```python
df.loc[df.duplicated(subset=['job_title'])].head(5)
```

```python
by_job_title = (df
            .groupby('job_title')
            .filter(lambda x : len(x) > 10)
            .groupby(['job_title', 'work_year'])
            .size()
            .unstack()
           )
#by_job_title=by_job_title.sort_values(by=[2023],ascending=False).head(10)
```

```python
plt.figure(figsize=(10,10))
g = sns.heatmap(
    by_job_title, 
    square=True, # make cells square
    cbar_kws={'fraction' : 0.01}, # shrink colour bar
    cmap='OrRd', # use orange/red colour map
    linewidth=1 # space between cells
)
```

```python
df2=df[df['work_year']==2023]
ax=df2['job_title'].value_counts()\
.head(10)\
.plot(kind='bar',title='Job Titles in 2023')
ax.set_xlabel('Job Titles')
ax.set_ylabel('count')
order=df['job_title'].value_counts(ascending=True)
for label in ax.containers:
    ax.bar_label(label)
plt.show()
```

```python
sns.set_style('ticks')
plt.figure(figsize=(10,10))
c=sns.countplot(x='experience_level',data=df2,hue='remote_ratio')
c.set_xlabel('Experience Level')
c.set_ylabel('Count')
for label in c.containers:
    c.bar_label(label)
plt.show()
```

```python
jobs = df[df['work_year']==2023]['job_title'].value_counts()[:10].reset_index()
fig, ax = plt.subplots()
ax = sns.barplot(ax = ax, data = jobs , y = jobs['index'], x = jobs.job_title)
ax.set(ylabel='Job Titles',xlabel='Counts', title='Top 10 Job Titles in 2023')
ax.bar_label(ax.containers[0], padding = 2)
```

```python
fig, ax = plt.subplots()
sns.countplot(ax = ax, data = df2, x = df.experience_level)
ax.set(xlabel='', ylabel='Counts', title='Experience Levels')
ax.bar_label(ax.containers[0])
```

```python
fig, ax = plt.subplots()
sns.countplot(ax = ax, data = df, x = df.employment_type, hue = 'experience_level')
ax.set(xlabel='Experience Level', ylabel='Counts', title='Number of Employment Types in 2023')
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
ax.bar_label(ax.containers[2])
ax.bar_label(ax.containers[3])
```

```python
df_job_title=df['salary_in_usd'].groupby(df['job_title']).mean().round(0).nlargest(15).sort_values(ascending=False).reset_index()
plt.figure(figsize=(25,9))
fig, ax = plt.subplots()
ax=sns.barplot(ax=ax,data=df_job_title,y=df_job_title.job_title,x=df_job_title.salary_in_usd)
ax.set(ylabel='Job titles',xlabel='Salary in usd', title='Top 15 Average Salaries by Job Titles')
ax.bar_label(ax.containers[0], padding = 2)
```

```python
df_employment_type=df.groupby(['employment_type', 'work_year'])['salary_in_usd'].mean().round(0).nlargest(15).sort_values(ascending=False).reset_index()
plt.figure(figsize=(10,10))
fig, ax = plt.subplots()
ax=sns.barplot(ax=ax,data=df_employment_type,x='employment_type',y='salary_in_usd',ci=None, hue='work_year')
ax.set(ylabel='Employment Type',xlabel='Salary in usd', title='Average Salaries by Employment Type Per Year')
ax.bar_label(ax.containers[0],padding=4)
ax.bar_label(ax.containers[1],padding=4)
ax.bar_label(ax.containers[2],padding=4)
ax.bar_label(ax.containers[3],padding=4)
```

```python
df_salary_per_year=df.groupby(['work_year'])['salary_in_usd'].mean().round(0).reset_index()
plt.figure(figsize=(25,9))
fig, ax = plt.subplots()
ax=sns.lineplot(ax=ax,data=df_salary_per_year,x=['2020','2021','2022','2023'],y='salary_in_usd')
ax.set(ylabel='Salary in usd',xlabel='Work Year', title='Average Salaries by Work Year')
```
